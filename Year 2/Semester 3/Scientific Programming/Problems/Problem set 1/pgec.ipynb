## Population Growth Dataset for each country(1961 - 2023)

This dataset contains population growth data (annual %) for all countries, sourced from the World Bank.

### Source
The original data can be found at [World Bank - Population Growth (Annual %)](https://data.worldbank.org/indicator/SP.POP.GROW).

### The data in this repo
Problem set 1/API_SP.POP.GROW_DS2_en_csv_v2_2594848.csv

### License

### Description
This dataset includes the following columns:
- **Country Name**
- **Country Code**
- **Year**
- **Population Growth (annual %)**
